```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
numerical = pd.read_csv("numerical.csv")
symbolic = pd.read_csv("categorical.csv")
both = pd.read_csv("both.csv")
```

```python
limit = 1000
```

```python
numerical = numerical.where(numerical.Instances > limit)
symbolic = symbolic.where(symbolic.Instances > limit)
both = both.where(both.Instances > limit)
```

```python
plt.figure(figsize=(7, 4), dpi=100)
ax = plt.axes()
# ax.set_facecolor('lightgray')

# the histogram of the data
plt.hist(np.log(both.Instances), bins=12, label="all datasets")
plt.hist(np.log(symbolic.Instances), bins=12, label="symbolic datasets")


plt.xlabel('size (log-scale)', fontdict={"size":12})
plt.title('Dataset size frequence', fontdict={"size":25})
plt.legend()
plt.grid(True)
plt.savefig("dataset-size-frequence.png")
plt.show()
```
